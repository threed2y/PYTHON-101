RECURSION USING PARAMETERS
-- HEAD RECURSION
---- 1 TO N 
---- N TO 1 
-- TAIL RECURSION
---- 1 TO N 
---- N TO 1

```python
# HEAD RECURSION
# PRINT X,N TIMES

def func(x,n):
    if n == 0:
        return

    print(x)
    func(x,n-1)
func(4,5)
```

```python
# PRINT 1 TO N

def func(x,n):
    if x > n:
        return

    print(x)
    func(x+1,n)

func(2,10)
       
```

```python
# PRINT N TO 1

def func(n):
    if n == 0:
        return

    print(n)
    func(n-1)

func(20)
```

```python
# TAIL RECURSION
# N TO 1

def func(i,n):
    if i>n:
        return

    func(i+1,n)
    print(i)

func(4,8)
```

```python
def func(x,n):
    if x == n:
        return

    func(x+1,n)  # DO IT BY BACKTRACING
    print(x)

func(1,8)
```
