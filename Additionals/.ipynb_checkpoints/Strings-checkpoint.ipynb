## EXTRACTION OF DIGITS
1. COUNT DIGITS
2. REVERSE A NUMBER
3. PALINDROME CHECK
4. ARMSTRONG NUMBER

```python
# 1. COUNT DIGITS

n = 5378 #NEVER MAKE CHANGES IN INPUT NUMBERS
num = n 
count = 0

while num > 0:
    r_num = num % 10
    count += 1 
    num = num //10

print(count)
```

```python
# 2. REVERSE A NUMBER

b = 4422
num_b = b

result = 0

while num_b > 0:
    ld = num_b % 10
    result = result*10 + ld 
    num_b = num_b // 10
print("reversed number: ",result)
## 3. Palindrome check
if result == b:
    print("palindrome")
else:
    print("not a palindrome")
```

```python
# 4. ARMSTRONG NUMBER -- 153 = 1^3 + 5^3 + 3^3

c = 153
cat = c
meow = 0

while cat > 0:
    cd = cat % 10
    meow += cd**3
    cat //= 10
print(meow)

if meow == c:
    print("armstrong number")
else:
    print("not a armstrong number")
    

```
