FINDING OUT A FACTOR
1. BRUTE FORCE
2. BETTER METHOD
3. OPTIMAL METHOD

```python
# 1. BRUTE FORCE
n = 25
factors = []

for i in range(1,n+1):
    if n % i == 0:
        factors.append(i)
print(factors)
        
```

```python
# 2. BETTER METHOD

c = 20
fac = []

for i in range(1,(c//2)+1): # DO IT TILL HALF OF THE NUMBER AND ADD THE NUMBER ITSELF TO THE LIST
    if c % i == 0:
        fac.append(i)
fac.append(c)
print(fac)
        
```

```python
# 3. BEST METHOD WITH TC -- O(sqrt_n)+o(nlogn) 

from math import sqrt
d = 25
factr = []

for i in range(1, int(sqrt(d))+1):
    if d % i == 0:
        factr.append(i)
        if d // i != i :
            factr.append(d//i)
factr.sort()
print(factr)
```
