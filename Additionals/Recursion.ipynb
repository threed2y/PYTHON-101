```python
def greet():
    print("hey hey hey")
greet()
```

```python
def greet():
    print("hey hey hey")
    greet()
greet()
```

```python
# HEAD RECURSION
count = 0
def func():
    if count == 4:
        return
    print("hey")
    count += 1
    func()
```

```python
# TAIL RECURSION
count = 0
def func():
    if count == 4:
        return
    count += 1
    func()
    print("batman")
    
```
