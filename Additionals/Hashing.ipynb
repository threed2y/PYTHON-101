HASHING
-- PRESTORING VALUES INTO SOME DATA STRUCTURE LIKE LIST/DICTIONARY/SETS AND THEN FETCHING IT.

```python
# 2. BRUTE FORCE
n = [5,3,2,2,1,5,5,7,5,10]
m = [10,11,1,9,5,6,7,2]

for i in n:
    count = 0
    for j in m:
        if i == j:
            count +=1
            
    print(count,i)
```

```python
# 3. HASHING THROUGH LIST
n = [5,3,2,2,1,5,5,7,5,10]
m = [10,11,1,9,5,6,7,2]

hash_list = [0]*11
for num in n:
    hash_list[num] += 1

for num in m:
    if num < 1 or num > 10:
        print(0)
    else:
        hash_list[num]
print(hash_list)
    
```

```python
# 4. HASHING THROUGH DICTIONARY
hash_map = {}

for i in range(len(n)):
    hash_map[n[i]] = hash_map.get(n[i],0)+1
print(hash_map)
```

```python
# 5. HASHING THROUGH CHARACTERS

hash_list = [0]* 27

s = "azyzyyzaa"
a = ["a","d","y","x"]

for chr in s:
    ascii_value = ord(chr)
    index = ascii_value - 97
    hash_list[index] += 1

for chr in a :
    ascii_value = ord(chr)
    index = ascii_value - 97
    print(hash_list[index])
```
