PARAMETRIC AND FUNCTIONIZED RECURSION
-- 1. PARAMETRIC RECURSION
-- 2. FUNCTIONAL RECURSION

```python
# 1. PARAMETRIC FUNCTION

def func(sum,i,n):
    if i > n:
        print(sum)
        return 
    func(sum+i,i+1,n)

func(0,5,8)

```

```python
# 2. FUNCTIONAL RECURSION

def func(n):
    if n == 1:
        return 1
    return n+func(n-1)

func(10)
```

```python
# FINDING FACTORIALS

def func(n):
    if n == 1:
        return 1

    return n*func(n-1)

func(5)
```
